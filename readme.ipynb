### data preparation
The input data should be a .dat file which consists of user information of a specific event. The users must be sorted in order by infected time. The dataset must contain at least following columns,  
'id', 'time_lapsed', 'favourites_count', 'listed_count',  'statuses_count', 'source_candidates', 'source_index', 'followers_count','friends_count','seed_index', 'generation', 'time_since_seed', 'user_created_days','normalized_statuses_count', 'normalized_followers_count','normalized_favourites_count', 'normalized_listed_count','normalized_friends_count'.  
 Here is an example of the structure of users dataset:  

 
 
If users' friend list and follower list are also collected, they should be stored in .dat file. The files should have 1 column, which named as friend_list(or follower_list). The index of the lists should corresponding to the users information dataset. (i.e. the friend list in the n-th row of the friend_list dataset should be the list of friend of user in the n-th row in the users dataset)
